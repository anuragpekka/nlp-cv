### Theoretical Probability

```python
def coin_flip_probability():
    head_prob = 1/2
    return head_prob

def rolling_dice_probability(number):
    if number not in range(1, 7):
        raise ValueError('Number must be between 1 and 6')
    probability = 1/6
    return probability
```

```python
coin_flip_result = coin_flip_probability()
coin_flip_result
```

```python
dice_roll_result = rolling_dice_probability(5)
dice_roll_result
```

### Experimental probability

```python
import random
import numpy as np
from collections import Counter
# simulating a coin flip experiment
def simulate_coin_flip_probs(num_flips):
    results = []
    for _ in range(num_flips):
        flip = random.choice(['heads','tails'])
        results.append(flip)
    counted = Counter(results)
    head_count = counted['heads']
    tail_count = counted['tails']
    head_prob = head_count/num_flips
    tail_prob = tail_count/num_flips
    return head_prob, tail_prob, results
```

```python
heads, tails, results = simulate_coin_flip_probs(1000)
```

```python
print(f"Total number of trials: {len(results)}")
print(f"Probability of heads: {heads}")
print(f"Probability of tails: {tails}")
```

### Linear growth

```python
x = np.arange(-10, 10, 0.25)
x
```

```python
y = np.pi* x + np.pi
y
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
```

```python
y0 = -np.pi* x + np.pi
y0
```

```python
plt.plot(x, y0)
```

### exponential growth


```python
yp = x**2
plt.plot(x, yp)
```

### logarithmic growth


```python
yl = np.log(x)
plt.plot(x, yl)
```

### Logistic growth(sigmoid)

```python
ys = (1 / (1 + (np.e**(-x)))) ## sigmoid
plt.plot(x, ys)
```
