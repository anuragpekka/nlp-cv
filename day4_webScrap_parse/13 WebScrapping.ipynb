```python
# all of the above code can be concised like this: ( can be copied anywhere)
url = urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
data = url.read()
soup  = BeautifulSoup(data, 'html.parser')
text = soup.get_text() # getting only the text
tokens = word_tokenize(text) # tokenizing the words
tokens1 = [token for token in tokens if token.lower() not in swords] # removing stopwords
tokens2 = [token.lower() for token in tokens1 if token.isalnum()] # only alphanumeric tokens will remain
freq = FreqDist(tokens2) # defining the frequency distribution plot
freq.plot(20)  # plotting the frequency distribution
freq.most_common(5) # most common words in the data
```

```python
# doing the same for tiger data, with a stemmer to avoid similar words(tiger, tigris, tigers in this situation) in the plot
from nltk.stem import LancasterStemmer, PorterStemmer, SnowballStemmer, WordNetLemmatizer
ls = LancasterStemmer()
ps = PorterStemmer()
ss = SnowballStemmer(language='english')
wn = WordNetLemmatizer()
url = urlopen('https://en.wikipedia.org/wiki/Tiger') # urllib
data = url.read() # read the data from the url
soup  = BeautifulSoup(data, 'html.parser')
text = soup.get_text() # getting only the text
tokens = word_tokenize(text) # tokenizing the words
tokens1 = [token for token in tokens if token.lower() not in swords] # removing stopwords
tokens2 = [token.lower() for token in tokens1 if token.isalnum()] # only alphanumeric tokens will remain
tokens3 = [wn.lemmatize(token) for token in tokens2 if len(token)>1] # test for different stemmers and lemmatizers here
freq = FreqDist(tokens3) # defining the frequency distribution plot
freq.plot(20) # plotting the 20 most frequent words
freq.most_common(5) # most common words in the data
```
