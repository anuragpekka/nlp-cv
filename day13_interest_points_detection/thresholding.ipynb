```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
## directories
images = '../images'
saved_images = '../saved_images'
trans_images = '../trans_images'
```

```python
img = cv2.imread(os.path.join(images, 'sachin.jpg'), 0)
```

#### Binary thresholding

```python
binary_image = img > 127
```

```python
binary_image
```

```python
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(binary_image, cmap='gray')
```

```python
lenna = cv2.imread(os.path.join(images, 'lenna.png'), 0)
binary_lenna = lenna > 127
new_lenna = binary_lenna.astype(np.uint8)*255 # simply changing the datatype

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')
plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(new_lenna, cmap='gray')
```

### Using cv2.threshold

```python
lenna = cv2.imread(os.path.join(images, 'lenna.png'),0)
thresh, new_lenna = cv2.threshold(lenna, 127, 255, cv2.THRESH_BINARY)

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')
plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(new_lenna, cmap='gray')
```

### Inverse binary thresholding(niggafication)

```python
lenna = cv2.imread(os.path.join(images, 'lenna.png'), 0)
binary_lenna = lenna < 127 # this is the change. Sign has been changed, compared to the prev code
new_lenna = binary_lenna.astype(np.uint8)*255 # simply changing the datatype

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')
plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(new_lenna, cmap='gray')
```

#### using cv2.threshold

```python
lenna = cv2.imread(os.path.join(images, 'lenna.png'), 0)
thresh, new_lenna = cv2.threshold(lenna, 127, 255, cv2.THRESH_BINARY_INV)

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')
plt.subplot(1,2,2)
plt.title('Binary Image')
plt.imshow(new_lenna, cmap='gray')
```

### Otsu thresholding

```python
lenna = cv2.imread(os.path.join(images, 'lenna.png'), 0)
thresh_otsu, lenna_otsu = cv2.threshold(lenna, 100, 255, cv2.THRESH_OTSU) 
thresh_bin, lenna_bin = cv2.threshold(lenna, 100, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(lenna, cmap='gray')
plt.subplot(1,3,2)
plt.title('Binary Image')
plt.imshow(lenna_bin, cmap='gray')
plt.subplot(1,3,3)
plt.title('Otsu Image')
plt.imshow(lenna_otsu, cmap='gray')
```
