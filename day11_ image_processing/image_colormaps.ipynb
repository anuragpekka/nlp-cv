## ColorMap
1. It's a predefined set of colors that are mapped to numerical values within an image.
2. It acts as a translator, converting numerical data into visually interpretable colors.
3. OpenCV offers a diverse palette of built-in colormaps.

```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
## images path
images = '../images/'
saved_images = '../saved_images/'
```

```python
img = cv2.imread(os.path.join(images,'sachin.jpg'), 0)
```

```python
## convert bgr to rgb

# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img, cmap='Blues')
```

### Custom Cmap creation

```python
orig_map = plt.colormaps['inferno']
reversed_map = orig_map.reversed()
plt.imshow(img, cmap=reversed_map)
```

### Experimenting with multiple such cmaps and their reversed images

```python
plt.figure(figsize=(16,8))
plt.subplot(2,5,1); plt.title('Blues')
plt.imshow(img, cmap=plt.colormaps['Blues'].reversed())

plt.subplot(2,5,2); plt.title('Greens')
plt.imshow(img, cmap=plt.colormaps['Greens'].reversed())

plt.subplot(2,5,3); plt.title('Reds')
plt.imshow(img, cmap=plt.colormaps['Reds'].reversed())

plt.subplot(2,5,4); plt.title('Oranges')
plt.imshow(img, cmap=plt.colormaps['Oranges'].reversed())

plt.subplot(2,5,5); plt.title('Greys')
plt.imshow(img, cmap=plt.colormaps['Greys'].reversed())

plt.subplot(2,5,6); plt.title('Purples')
plt.imshow(img, cmap=plt.colormaps['Purples'].reversed())

plt.subplot(2,5,7); plt.title('YlOrBr')
plt.imshow(img, cmap=plt.colormaps['YlOrBr'].reversed())

plt.subplot(2,5,8); plt.title('OrRd')
plt.imshow(img, cmap=plt.colormaps['OrRd'].reversed())

plt.subplot(2,5,9); plt.title('rainbow real')
plt.imshow(img, cmap=plt.colormaps['rainbow'])

plt.subplot(2,5,10); plt.title('rainbow reversed')
plt.imshow(img, cmap=plt.colormaps['rainbow'].reversed())
plt.show()
```

### Diverging colormaps
Coolwarm, seismic, 


```python
plt.figure(figsize=(16,8))

plt.subplot(2,3,1); plt.title('coolwarm')
plt.imshow(img, cmap=plt.colormaps['coolwarm'])

plt.subplot(2,3,4); plt.title('coolwarm reversed')
plt.imshow(img, cmap=plt.colormaps['coolwarm'].reversed())

plt.subplot(2,3,2); plt.title('seismic')
plt.imshow(img, cmap=plt.colormaps['seismic'])

plt.subplot(2,3,5); plt.title('seismic reversed')
plt.imshow(img, cmap=plt.colormaps['seismic'].reversed())

plt.subplot(2,3,3); plt.title('PiYG')
plt.imshow(img, cmap='PiYG')

plt.subplot(2,3,6); plt.title('PiYG reversed')
plt.imshow(img, cmap=plt.colormaps['PiYG'].reversed())
```

### Cyclic colormaps
twilight, twilight_shifted, hsv

```python
plt.figure(figsize=(16,8))

plt.subplot(2,4,1); plt.title('twilight')
plt.imshow(img, cmap=plt.colormaps['twilight'])

plt.subplot(2,4,5); plt.title('twilight reversed')
plt.imshow(img, cmap=plt.colormaps['twilight'].reversed())

plt.subplot(2,4,2); plt.title('twilight_shifted')
plt.imshow(img, cmap=plt.colormaps['twilight_shifted'])

plt.subplot(2,4,6); plt.title('twilight_shifted reversed')
plt.imshow(img, cmap=plt.colormaps['twilight_shifted'].reversed())

plt.subplot(2,4,3); plt.title('hsv')
plt.imshow(img, cmap='hsv')

plt.subplot(2,4,7); plt.title('hsv reversed')
plt.imshow(img, cmap=plt.colormaps['hsv'].reversed())

plt.subplot(2,4,4); plt.title('plasma')
plt.imshow(img, cmap='plasma')

plt.subplot(2,4,8); plt.title('plasma reversed')
plt.imshow(img, cmap=plt.colormaps['plasma'].reversed())
```
