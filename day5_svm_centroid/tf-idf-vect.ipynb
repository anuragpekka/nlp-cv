```python
from nltk.tokenize import word_tokenize
import pandas as pd
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
sentence1 = 'It is a good practice for us.'
sentence2 = 'It was also good to know about it.'
```

```python
x_new = tfidf.fit_transform([sentence1, sentence2])
x_new.toarray()
```

```python
df_tfidf = pd.DataFrame(data=x_new.toarray(), columns=tfidf.get_feature_names_out())
df_tfidf
```

```python
# Opening a file with multiple sentences and tf-idf on it
tf_idf_vect = TfidfVectorizer()
with open('doc1.txt', 'r') as f:
    tf_idf_array = tf_idf_vect.fit_transform(f.readlines()).toarray() # fit_transform accepts a list of sentences and then the sparse matrix is converted to numpy array
tf_idf_array
```

```python
df_new = pd.DataFrame(data=tf_idf_array, columns=tf_idf_vect.get_feature_names_out())
df_new
```
