```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import nltk
```

### Punctuation removal

```python
text = 'HELLO Brother!!!!!! How are you? Welcome to Pune.'
```

```python
import string
tokens = [word for word in word_tokenize(text) if word not in string.punctuation]
tokens
```

```python
' '.join(tokens)
```

## Stopwords Removal

```python
swords = stopwords.words('english') # stopwords
```

```python
swords
```

```python
tokens_cleaned = [word for word in tokens if word.lower() not in swords] # don't forget to include .lower() in the condition
tokens_cleaned
```

```python
text_cleaned = ' '.join(tokens_cleaned)  #keywords after removing the stopwords
text_cleaned
```

```python
# swords in nepali
swords_nepali = stopwords.words('nepali')
swords_nepali
```

```python
nepali_text = 'नेपाल एउटा सुन्दर देश हो जहाँ हिमालहरू र पहाडहरू छन्. हामी प्रत्येक बिहान सूर्य उदाउँदा रमाइलो अनुभव गर्छौं'
```

```python
# filtering the stopwords
nepali_cleaned = [word for word in word_tokenize(nepali_text) if word not in swords_nepali]
nepali_cleaned
```

```python
' '.join(nepali_cleaned)
```
